```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(1000)

plt.hist(data, bins=10)
plt.title("Histogram")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

```python
import matplotlib.pyplot as plt

days = [1, 2, 3, 4, 5]
temperature = [30, 32, 31, 29, 35]

plt.plot(days, temperature)
plt.title("Line Chart")
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [5, 20, 15, 25, 30]

plt.scatter(x, y)
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt

sizes = [25, 30, 20, 25]
labels = ["A", "B", "C", "D"]

plt.pie(sizes, labels=labels, autopct="%1.1f%%")
plt.title("Pie Chart")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 17, 20]

plt.fill_between(x, y)
plt.title("Area Chart")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix

df = pd.DataFrame({
    "A": [1, 2, 3, 4, 5],
    "B": [5, 3, 4, 2, 1],
    "C": [10, 20, 15, 25, 30]
})

scatter_matrix(df, figsize=(8, 8))
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
tips=sns.load_dataset('tips')
tips.head()
sns.scatterplot(data=tips, x= 'total_bill', y = 'tip',hue='sex')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
tips=sns.load_dataset('tips')
tips.head()
sns.barplot(data=tips, x= 'day', y = 'total_bill', hue='sex')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
tips=sns.load_dataset('tips')
tips.head()
sns.boxplot(data=tips, x= 'day', y = 'total_bill', hue='sex')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
tips=sns.load_dataset('tips')
tips.head()
sns.violinplot(data=tips, x= 'day', y = 'total_bill', hue='sex',split=True)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
tips=sns.load_dataset('tips')
tips.head()
sns.stripplot(data=tips, x= 'day', y = 'total_bill', hue='sex',jitter=False)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
tips=sns.load_dataset('tips')
tips.head()
sns.kdeplot(data=tips ['total_bill'], fill=True)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
tips=sns.load_dataset('tips')
tips.head()
sns.regplot(data=tips, x= 'total_bill', y = 'tip')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
tips=sns.load_dataset('tips')
tips.head()
sns.lmplot(data=tips, x= 'total_bill', y = 'tip', hue='sex')
plt.xlabel('total_bill')
plt.ylabel('tip')
plt.title('total bill vs tips')
sns.set_palette('bright')

plt.grid(True)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
tips=sns.load_dataset('tips')
tips.head()
sns.displot(data=tips, x= 'total_bill', hue='sex', kind='kde',fill=True,palette='coolwarm', aspect=1.5)
plt.xlabel('total_bill')
plt.ylabel('tip')
plt.title('total bill vs tips')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
tips=sns.load_dataset('tips')
tips.head()
sns.histplot(data=tips, x= 'total_bill', hue='sex', multiple='dodge')
plt.show()
```
